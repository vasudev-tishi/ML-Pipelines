**2.Implement a Bagging classifier with Decision Trees.**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score
```

```python
df=pd.read_csv('/content/classification_dataset.csv')
```

```python
df.head()
```

```python
# Split feauters and labels
X=df.drop("label",axis=1)
y=df["label"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
base_tree = DecisionTreeClassifier(random_state=3)
```

```python
bagging_clf = BaggingClassifier(
        estimator=base_tree,
        n_estimators=50,
        random_state=3,
        n_jobs=-1
    )
```

```python
bagging_clf.fit(X_train,y_train)
```

```python
y_pred=bagging_clf.predict(X_test)
```

```python
print("Accuracy",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

**3.Compare its performance with Random Forest and AdaBoost.**

```python
rf_clf = RandomForestClassifier(n_estimators=50, random_state=42, n_jobs=-1)
ada_clf = AdaBoostClassifier(estimator=base_tree, n_estimators=50, random_state=42)
```

```python
models={"Bagging":bagging_clf,
        "RandomForest":rf_clf,
        "AdaBoost":ada_clf}
```

```python
results={}
```

```python
for name,model in models.items():
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  results[name]={
      "accuracy":accuracy_score(y_test,y_pred),
      "report":classification_report(y_test,y_pred,output_dict=False)
  }
```

```python
for name, metrics in results.items():
    print("="*40)
    print(f"{name} Classifier")
    print("Accuracy:", metrics["accuracy"])
    print(metrics["report"])
```

4. Report accuracy, precision, and recall for each method, and briefly explain which

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results[name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred)
    }

# Show results
report_df = pd.DataFrame(results).T
print(report_df)
```
